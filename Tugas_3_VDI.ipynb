```python
#Visualisasi dengan Pie Chart
import matplotlib.pyplot as plt
import seaborn as sns

# Menghitung tag/genre yang paling umum dari kolom 'Tags'
steam_data['Tags_list'] = steam_data['Tags'].fillna('').apply(lambda x: x.split(','))

# Flatten list of tags
tags_flat = [tag.strip() for sublist in steam_data['Tags_list'] for tag in sublist]

# Hitung kemunculan tiap tag/genre
tag_counts = pd.Series(tags_flat).value_counts().head(10)  # Top 10 genre yang paling umum

# Tentukan warna manual untuk masing-masing genre
colors = ['#FF6666','#6666FF','#66CC66','#FFCCE5','#FFD966','#CC99FF','#FF9999','#6666CC','#99FF99','#CCCCCC']


# Plot pie chart dengan warna manual
plt.figure(figsize=(8, 8))
plt.pie(tag_counts, labels=tag_counts.index, autopct='%1.1f%%',
        colors=colors, startangle=140, wedgeprops={'edgecolor': 'gray'})

# Tambahkan judul
plt.title('Proporsi Genre Game Paling Populer di Steam', fontsize=16)

# Tampilkan pie chart
plt.tight_layout()
plt.show()
```

Visualisasi Pie Chart ini membantu memberikan gambaran pada pemain bagaimana genre dari game terdistribusi dalam Steam, dan juga bisa digunakan sebagai pemahaman tres pasar dan minat para gamer.
